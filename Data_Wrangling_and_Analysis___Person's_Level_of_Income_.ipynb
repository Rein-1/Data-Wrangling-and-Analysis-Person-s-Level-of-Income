```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
data = pd.read_csv(url, header=None)

data.head()
```

```python
column_names = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',
                'marital_status', 'occupation', 'relationship', 'race', 'sex',
                'capital_gain', 'capital_loss', 'hours_per_week', 'native_country',
                'income']
data.columns = column_names

data
```

```python
data['income'] = data['income'].apply(lambda x: 1 if x == ' >50K' else 0)
data
```

```python
high_income = data[data['income'] == 1]
high_income
```

```python
low_income = data[data['income'] == 0]
low_income
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='marital_status', hue='income', data=data, palette='RdBu')
plt.title("Marital Status vs Income", fontsize=16)
plt.xlabel("Marital Status", fontsize=14)
plt.ylabel("Count", fontsize=14)
plt.xticks(rotation=45, ha='right', fontsize=12)

plt.show()
```

```python
counts = data['income'].value_counts()
percentages = [counts[0]/len(data)*100, counts[1]/len(data)*100]

labels = ['<=50K', '>50K']
colors = ['red', 'navy']
plt.pie(percentages, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Percentage of Incomes')

plt.axis('equal')
plt.show()
```

```python
sns.histplot(x='age', data=data)
```

```python
corr_matrix = data.corr()

fig, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f',
            xticklabels=['Age', 'Final Weight', 'Years of Education', 'Capital Gains',
                         'Capital Losses', 'Hours per Week'],
            yticklabels=['Age', 'Final Weight', 'Years of Education', 'Capital Gains',
                         'Capital Losses', 'Hours per Week'])

ax.set_title('Correlation Matrix Heatmap')
```

```python
sns.countplot(x='income', data=data)
plt.show()
```

```python
sns.histplot(x="capital_loss", data=data, kde=True)
plt.title("Distribution of Capital Loss")
plt.xlabel("Capital Loss")
plt.ylabel("Count")
plt.show()
```

```python
import numpy as np
median_age = np.median(data['age'])
print("Median Age:", median_age)
```

```python
high_income_data = data[data['income'] == '>50K']
mean_education_level = np.mean(high_income_data['education_num'])
print("Mean Education Level (Income >50K):", mean_education_level)
```

```python
capital_gains = data['capital_gain']
num_greater_than_0 = np.count_nonzero(capital_gains)
print("Number of Individuals with Capital Gain >0:", num_greater_than_0)
```
